```python
# Solution - enter your code solution below
import tkinter as tk
from tkinter import messagebox
from tkinter import ttk
import calendar
from datetime import datetime

class Caregiver:
    def __init__(self, name, phone, email, pay_rate=20):
        self.name = name
        self.phone = phone
        self.email = email
        self.pay_rate = pay_rate
        self.availability = {
            '7:00 AM - 1:00 PM': 'available',  # Default status is 'available'
            '1:00 PM - 7:00 PM': 'available'
        }
        self.hours_worked = 0  # Total hours worked by the caregiver

    def update_availability(self, shift, status):
        """Update availability for a particular shift."""
        if shift in self.availability:
            self.availability[shift] = status

    def log_hours(self, hours):
        """Log hours worked by the caregiver."""
        self.hours_worked += hours

    def calculate_weekly_pay(self):
        """Calculate weekly pay based on hours worked."""
        return self.hours_worked * self.pay_rate

    def __str__(self):
        return f"{self.name} - {self.phone} - {self.email}"

class Scheduler:
    def __init__(self, caregivers):
        self.caregivers = caregivers
        self.schedule = {}  # Key: Date, Value: Shift assignments (AM, PM)

    def assign_shift(self, date, shift, caregiver):
        """Assign a caregiver to a shift on a particular date."""
        if caregiver.availability.get(shift) in ['preferred', 'available']:
            if date not in self.schedule:
                self.schedule[date] = {}
            self.schedule[date][shift] = caregiver.name
            caregiver.log_hours(6)  # Each shift is 6 hours
        else:
            print(f"{caregiver.name} is unavailable for the {shift} on {date}.")

    def display_schedule(self):
        """Display the schedule in a readable format."""
        for date, shifts in self.schedule.items():
            print(f"Schedule for {date}:")
            for shift, caregiver in shifts.items():
                print(f"  {shift}: {caregiver}")
            print("\n")

class PayReport:
    def __init__(self, caregivers):
        self.caregivers = caregivers

    def generate_report(self):
        report = "Weekly Pay Report\n"
        total_monthly_pay = 0
        for caregiver in self.caregivers:
            weekly_pay = caregiver.calculate_weekly_pay()
            total_monthly_pay += weekly_pay
            report += f"{caregiver.name}: ${weekly_pay}\n"
        report += f"\nTotal Monthly Pay: ${total_monthly_pay}"
        return report

class CaregiverSchedulerApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Caregiver Scheduler")
        
        self.caregivers = []  # List to hold caregiver objects
        self.scheduler = Scheduler(self.caregivers)
        
        # Caregiver details entry frame
        self.caregiver_frame = tk.LabelFrame(self.root, text="Caregiver Details", padx=10, pady=10)
        self.caregiver_frame.grid(row=0, column=0, padx=10, pady=10, sticky="w")

        # Caregiver name
        self.name_label = tk.Label(self.caregiver_frame, text="Name:")
        self.name_label.grid(row=0, column=0, sticky="w")
        self.name_entry = tk.Entry(self.caregiver_frame)
        self.name_entry.grid(row=0, column=1)

        # Caregiver phone
        self.phone_label = tk.Label(self.caregiver_frame, text="Phone:")
        self.phone_label.grid(row=1, column=0, sticky="w")
        self.phone_entry = tk.Entry(self.caregiver_frame)
        self.phone_entry.grid(row=1, column=1)

        # Caregiver email
        self.email_label = tk.Label(self.caregiver_frame, text="Email:")
        self.email_label.grid(row=2, column=0, sticky="w")
        self.email_entry = tk.Entry(self.caregiver_frame)
        self.email_entry.grid(row=2, column=1)

        # Add caregiver button
        self.add_button = tk.Button(self.caregiver_frame, text="Add Caregiver", command=self.add_caregiver)
        self.add_button.grid(row=3, column=0, columnspan=2)

        # Shift assignment frame
        self.shift_frame = tk.LabelFrame(self.root, text="Assign Shifts", padx=10, pady=10)
        self.shift_frame.grid(row=1, column=0, padx=10, pady=10, sticky="w")

        # Date and Shift selection
        self.date_label = tk.Label(self.shift_frame, text="Date (YYYY-MM-DD):")
        self.date_label.grid(row=0, column=0, sticky="w")
        self.date_entry = tk.Entry(self.shift_frame)
        self.date_entry.grid(row=0, column=1)

        self.shift_label = tk.Label(self.shift_frame, text="Shift:")
        self.shift_label.grid(row=1, column=0, sticky="w")
        self.shift_combo = ttk.Combobox(self.shift_frame, values=["7:00 AM - 1:00 PM", "1:00 PM - 7:00 PM"])
        self.shift_combo.grid(row=1, column=1)
        self.shift_combo.set("7:00 AM - 1:00 PM")  # Default shift

        self.caregiver_label = tk.Label(self.shift_frame, text="Caregiver:")
        self.caregiver_label.grid(row=2, column=0, sticky="w")
        self.caregiver_combo = ttk.Combobox(self.shift_frame)
        self.caregiver_combo.grid(row=2, column=1)

        self.assign_button = tk.Button(self.shift_frame, text="Assign Shift", command=self.assign_shift)
        self.assign_button.grid(row=3, column=0, columnspan=2)

        # Pay Report button
        self.pay_button = tk.Button(self.root, text="Generate Pay Report", command=self.generate_pay_report)
        self.pay_button.grid(row=2, column=0, padx=10, pady=10)

        # Calendar Frame
        self.calendar_frame = tk.LabelFrame(self.root, text="Calendar View", padx=10, pady=10)
        self.calendar_frame.grid(row=3, column=0, padx=10, pady=10, sticky="w")

    def add_caregiver(self):
        """Add caregiver to the system."""
        name = self.name_entry.get()
        phone = self.phone_entry.get()
        email = self.email_entry.get()
        if name and phone and email:
            caregiver = Caregiver(name, phone, email)
            self.caregivers.append(caregiver)
            self.caregiver_combo['values'] = [c.name for c in self.caregivers]
            self.name_entry.delete(0, tk.END)
            self.phone_entry.delete(0, tk.END)
            self.email_entry.delete(0, tk.END)
            messagebox.showinfo("Success", f"Caregiver {name} added!")
        else:
            messagebox.showwarning("Input Error", "Please fill in all fields.")

    def assign_shift(self):
        """Assign selected caregiver to the chosen shift on the selected date."""
        date = self.date_entry.get()
        shift = self.shift_combo.get()
        caregiver_name = self.caregiver_combo.get()

        caregiver = next((c for c in self.caregivers if c.name == caregiver_name), None)
        if caregiver:
            self.scheduler.assign_shift(date, shift, caregiver)
            messagebox.showinfo("Success", f"Assigned {caregiver_name} to {shift} on {date}.")
        else:
            messagebox.showwarning("Input Error", "Caregiver not found.")

    def generate_pay_report(self):
        """Generate and show the pay report."""
        pay_report = PayReport(self.caregivers)
        report = pay_report.generate_report()
        messagebox.showinfo("Pay Report", report)


if __name__ == "__main__":
    root = tk.Tk()
    app = CaregiverSchedulerApp(root)
    root.mainloop()
```
